```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Define all functions
def translation(a, b, img):
    if len(img.shape) == 2:  
        height, width = img.shape
    else: 
        height, width, channels = img.shape
    
    T = np.float32([[1, 0, a], [0, 1, b]])
    img_translation = cv2.warpAffine(img, T, (width, height))
    plt.imshow(cv2.cvtColor(img_translation, cv2.COLOR_BGR2RGB) 
               if len(img.shape) == 3
               else img_translation, cmap="gray")
    
    plt.title('Translation')
    plt.show()
    return img_translation

def rotation(angle, scale, img):
    if len(img.shape) == 2:  
        height, width = img.shape
    else: 
        height, width, channels = img.shape
    
    center = (width / 2, height / 2)
    M = cv2.getRotationMatrix2D(center, angle, scale)
    img_rotation = cv2.warpAffine(img, M, (width, height))
    plt.imshow(cv2.cvtColor(img_rotation, cv2.COLOR_BGR2RGB) if len(img.shape) == 3 else img_rotation, cmap="gray")
    plt.title('Rotation')
    plt.show()
    return img_rotation

def scaling(sx, sy, img):
    img_resized = cv2.resize(img, None, fx=sx, fy=sy)
    plt.imshow(cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB) if len(img.shape) == 3 else img_resized, cmap="gray")
    plt.title('Scaling')
    plt.show()
    return img_resized

def reflection(axis, img):
    if axis == 0:  
        img_reflection = cv2.flip(img, 0)
    elif axis == 1: 
        img_reflection = cv2.flip(img, 1)
    else:
        print("Invalid axis for reflection")
        return img
    
    plt.imshow(cv2.cvtColor(img_reflection, cv2.COLOR_BGR2RGB) if len(img.shape) == 3 else img_reflection, cmap="gray")
    plt.title('Reflection')
    plt.show()
    return img_reflection

def shearing(axis, img):
    if len(img.shape) == 2:  
        height, width = img.shape
    else: 
        height, width, channels = img.shape

    if axis == 0: 
        M = np.float32([[1, 0.5, 0], [0, 1, 0]])
    elif axis == 1:  
        M = np.float32([[1, 0, 0], [0.5, 1, 0]])
    else:
        print("Invalid axis for shearing")
        return img
    
    img_sheared = cv2.warpAffine(img, M, (width, height))
    plt.imshow(cv2.cvtColor(img_sheared, cv2.COLOR_BGR2RGB) if len(img.shape) == 3 else img_sheared, cmap="gray")
    plt.title('Shearing')
    plt.show()
    return img_sheared

# Test Functions
def test_translation(img):
    tx, ty = 100, 50  # Example translation values
    return translation(tx, ty, img)

def test_rotation(img):
    angle, scale = 45, 1.0  # Example rotation angle and scale
    return rotation(angle, scale, img)

def test_scaling(img):
    sx, sy = 1.5, 1.5  # Example scaling factors
    return scaling(sx, sy, img)

def test_reflection(img):
    axis = 1  # Example reflection: 0 for x-axis, 1 for y-axis
    return reflection(axis, img)

def test_shearing(img):
    axis = 0  # Example shearing: 0 for x-axis, 1 for y-axis
    return shearing(axis, img)

# Main Script
img_path = "tulips.jpg"  # Replace with your image path
color_choice = 2  # 1 for grayscale, 2 for color

# Load image
if color_choice == 1:
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
elif color_choice == 2:
    img = cv2.imread(img_path, cv2.IMREAD_COLOR)
else:
    raise ValueError("Invalid color choice. Use 1 for grayscale or 2 for color.")

if img is None:
    raise FileNotFoundError("Image not found. Check the path and try again.")

# Apply transformations
print("Testing Translation...")
test_translation(img)

print("Testing Rotation...")
test_rotation(img)

print("Testing Scaling...")
test_scaling(img)

print("Testing Reflection...")
test_reflection(img)

print("Testing Shearing...")
test_shearing(img)
```
