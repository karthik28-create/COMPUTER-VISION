```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
img_src=cv2.imread("tulips.jpg")
plt.imshow(img_src)
```

```python
def translation(tx, ty, img_src):
    T = np.float32([[1, 0, tx], [0, 1, ty]])
    h, w = img_src.shape[:2]  # Corrected to get height and width correctly
    img_dst = cv2.warpAffine(img_src, T, (w, h))
    return img_dst

# Assuming img_src is already defined and loaded
T_img = translation(200, 300, img_src)

# Convert the image from BGR to RGB for correct color display in matplotlib
plt.imshow(cv2.cvtColor(T_img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
def Rotation(an, img_src, sf):
    h, w = img_src.shape[:2]  # Corrected to get height and width correctly
    cr = (w // 2, h // 2)  # Center of rotation
    RM = cv2.getRotationMatrix2D(cr, an, sf)  # Added scaling factor to the rotation matrix
    img_dst = cv2.warpAffine(img_src, RM, (w, h))
    return img_dst

# Assuming img_src is already defined and loaded
rotated_img = Rotation(-60, img_src, 0.6)

# Convert the image from BGR to RGB for correct color display in matplotlib
plt.imshow(cv2.cvtColor(rotated_img, cv2.COLOR_BGR2RGB))
plt.show()
```
